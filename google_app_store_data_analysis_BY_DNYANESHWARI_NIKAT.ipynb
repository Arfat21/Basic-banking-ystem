Next is reviews, review sometime can measure the app popularity. The more reviews, the better.

```python
# Reviews
df.Reviews.unique()
```

```python
# inside review there is a value with 3.0M with M stand for million, lets change it so it can be measure as float

Reviews = []

for x in df.Reviews:
    x = x.replace('M','00')
    Reviews.append(x)

Reviews = list(map(float, Reviews))
df['reviews'] = Reviews
sns.distplot(Reviews)
```

```python
g = sns.FacetGrid(df, col='Category', palette="Set1",  col_wrap=5, height=4)
g = (g.map(plt.hist, "Reviews", color="g"))
```

```python
# Total reviews
plt.figure(figsize=(12,12))
sum_rew = df.groupby(['Category'])['Reviews'].sum().sort_values(ascending=False)
sns.barplot(x=sum_rew, y=sum_rew.index, data=df)
```

showingn the total amount of total review

```python
plt.figure(figsize=(12,12))
mean_rew = df.groupby(['Category'])['Reviews'].mean().sort_values(ascending=False)
sns.barplot(x=mean_rew, y=mean_rew.index, data=df)
```

this is the average of reviews on each category lrts see the column ' Installs'

```python
# Installs
df.Installs.unique()
```

```python
df['Installs'].replace(to_replace=['0', 'Free'], value=['0+','0+'],inplace=True)
Installs = []

for x in df.Installs:
    x = x.replace(',', '')
    Installs.append(x[:-1])

Installs = list(map(float, Installs))
df['installs'] = Installs
sns.distplot(Installs)
```

distributed value of Installs on each category 

```python
g = sns.FacetGrid(df, col='Category', palette="Set1",  col_wrap=5, height=4)
g = (g.map(plt.hist, "Installs", bins=5, color='c'))
```

```python
# Total Installs
plt.figure(figsize=(12,12))
sum_inst = df.groupby(['Category'])['Installs'].sum().sort_values(ascending=False)
sns.barplot(x=sum_inst, y=sum_inst.index, data=df)
```

```python
#Lets is the version of android you should have before accessing the app.


# Android Version

df['Minimum Android Ver'].unique()
```

```python
#Now I am going to group it to 1 till 8 version of android. Change the null value to 1.0.

1
df['Minimum Android Ver'].replace(to_replace=['4.4W and up','Varies with device'], value=['4.4','1.0'],inplace=True)
2
df['Minimum Android Ver'].replace({k: '1.0' for k in ['1.0','1.0 and up','1.5 and up','1.6 and up']},inplace=True)
3
df['Minimum Android Ver'].replace({k: '2.0' for k in ['2.0 and up','2.0.1 and up','2.1 and up','2.2 and up','2.2 - 7.1.1','2.3 and up','2.3.3 and up']},inplace=True)
4
df['Minimum Android Ver'].replace({k: '3.0' for k in ['3.0 and up','3.1 and up','3.2 and up']},inplace=True)
5
df['Minimum Android Ver'].replace({k: '4.0' for k in ['4.0 and up','4.0.3 and up','4.0.3 - 7.1.1','4.1 and up','4.1 - 7.1.1','4.2 and up','4.3 and up','4.4','4.4 and up']},inplace=True)
6
df['Minimum Android Ver'].replace({k: '5.0' for k in ['5.0 - 6.0','5.0 - 7.1.1','5.0 - 8.0','5.0 and up','5.1 and up']},inplace=True)
7
df['Minimum Android Ver'].replace({k: '6.0' for k in ['6.0 and up']},inplace=True)
8
df['Minimum Android Ver'].replace({k: '7.0' for k in ['7.0 - 7.1.1','7.0 and up','7.1 and up']},inplace=True)
9
df['Minimum Android Ver'].replace({k: '8.0' for k in ['8.0 and up']},inplace=True)
10
df['Minimum Android Ver'].fillna('1.0', inplace=True)
```

```python
print(df.groupby('Category')['Minimum Android Ver'].value_counts())
Type_cat = df.groupby('Category')['Minimum Android Ver'].value_counts().unstack().plot.barh(figsize=(15,20), width=3)
plt.show()
```
